```python
# Import dependencies.
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using MATLAB
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() 
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# To create a quick graph using the object-oriented interface method
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Use plt.show()
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#5.1.4 Annotate Charts
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()

# Question 1 - use xlim() to show the months January through June on the graph
# plt.xlim("Jan", "June")
# plt.xlim(0, 5)
```

```python
# 5.1.4 - Skill Drill - Using the object-oriented approach, make the following changes to your line chart
# Change the line color to green.
# Change the marker to a diamond.
# Change the line width to 2 points.
# Add a legend for the city of Boston.
# Add a title and axes labels.
# Add a y-axis limit.
# Add grid lines.
ax = plt.axes()
ax.plot(x_axis, y_axis,  color='green', marker='d', linewidth=2, label='Boston')
ax.legend()
ax.set_title("PyBer Fare by Month") 
ax.set_xlabel('Date')
ax.set_ylabel('Fare ($)')
ax.set_ylim(0,45)
ax.grid()
```

```python
# Skill Drill Using the object-oriented approach, make the following changes to your bubble chart:
fig, ax = plt.subplots()
# 1. Change the color of the markers to sky bluel
# 2. Change the size of the markers to 5 times each data point.
# 3. Make the color 20% transparent.
# 4. Add a black edge color to the circles.
# 8. Switch the axis so that the Fare($) data is on the x-axis.
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", alpha=0.8, edgecolors="black", linewidths=2, label='Boston')
# 5. Make the linewidth of the circles 2 points.
# 6. Add a legend for the city of Boston.
ax.legend()
# 7. Add a title.
ax.set_title("PyBer Fare by Month") 
ax.set_xlabel('Fare ($)')
ax.set_ylabel('Date')
# 9. Add a limit for the x-axis data.
ax.set_xlim(0, 50)
# 10. Invert the y-axis so that January is at the top.
ax.invert_yaxis()
```

```python
# 5.1.8 Create Pie Charts
# Create a Pie Chart Using the MATLAB Method
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# add percentages for each month and "explode" the largest percentage, which is July, the seventh value in the x_axis
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
