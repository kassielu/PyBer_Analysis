```python
import matplotlib.pyplot as plt
```

```python
categories = ['pants', 'shirts','outerwear', 'pajamas', 'underwear', 'shoes']
revenue = [25,15,5,20,10,25]
```

```python
plt.pie(revenue)
plt.show
```

```python
plt.pie(revenue,labels=categories)
plt.show()
```

```python
plt.pie(revenue,labels=categories,autopct="%1.1f%%")
plt.show()
```

```python
explode = [0,.2,0,0,0,0]
```

```python
plt.pie(revenue,labels=categories,autopct="%1.1f%%",explode=explode)
plt.show()
```
